```python
#dropping the rows containing null values
df = df.dropna()
```

```python
#checking for null values again
df.isnull().sum()
```

```python
#displaying the columns of dataset
df.columns
```

```python
#replacing the index numbers with ids
df.set_index("id", inplace = True)
```

```python
df.head()
```

```python
#rename sqft_living and sqft_lot column name
df.rename(columns = {"sqft_living" : "living", "sqft_lot" : "lot"}, inplace = True)
```

```python
#determining the unique values of bathrooms column
df.bathrooms.unique()
```

```python
#finding max value of floors column
df.floors.max()
```

```python
#displaying only price, bedrooms, and living columns
df[['price','bedrooms','living']]
```

```python
#one-hot encoding of the bedroos column
dummies = pd.get_dummies(df.bedrooms)
```

```python
dummies.head()
```

```python
#concatenation of two dataframes df and dummies
df = pd.concat([df,dummies],axis='columns')
```

```python
df.head()
```

```python
#dropping the unwanted bedroom column
df.drop(["bedrooms"], axis='columns', inplace= True)
```
